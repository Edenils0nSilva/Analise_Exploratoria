````python
sns.scatterplot(data=iris, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
# Hipótese 2: O comprimento da pétala é um bom indicador para diferenciar a íris-virginica das outras espécies.
# Hipótese 6: A íris-virginica tem a maior variabilidade no comprimento da pétala entre as três espécies.
````

````python
# Calcular a média do comprimento das sépalas para cada espécie
mean_sepal_length = iris.groupby('Species')['SepalLengthCm'].mean()

# Criar um barplot
plt.figure(figsize=(8, 6))
sns.barplot(x=mean_sepal_length.index, y=mean_sepal_length.values)
plt.title('Média do Comprimento das Sépalas por Espécie')
plt.xlabel('Espécie')
plt.ylabel('Média do Comprimento das Sépalas')
plt.show()

# Hipótese 3: As íris-versicolor têm, em média, sépalas mais longas do que as íris-setosa.
````

````python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris, hue='Species')
plt.title('Boxplot de Comprimento da Pétala por Espécie')
````

````python
sns.histplot(data=iris, x='PetalLengthCm', hue='Species', kde=True)
plt.title('Distribuição do Comprimento das Pétalas por Espécie')
plt.show()
````

````python
sns.FacetGrid(iris, hue="Species", height=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
plt.title('Scatterplot de Comprimento da Sépala vs Largura da Sépala por Espécie')

#Hipótese 1: A largura da sépala é maior nas íris-setosa do que nas outras espécies.
````

````python
plt.figure(figsize=(12, 6))  # Altere o tamanho da figura aqui
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, hue="Species")
plt.show()
````
